```python
import pandas as pd

atlite_data = pd.read_csv('C:/Veda/VerveStacks/data/hourly_profiles/atlite_grid_cell_2013.csv')
```

```python
import pandas as pd

sarah_iso_data = pd.read_csv('C:/Veda/VerveStacks/data/hourly_profiles/sarah_era5_iso_2013.csv')

iso_list = atlite_data['grid_cell'].str[:3].unique()

sarah_iso_data = sarah_iso_data[sarah_iso_data['iso'].isin(iso_list)]

rezoning_solar_data = pd.read_csv('C:/Veda/VerveStacks/data/REZoning/REZoning_solar.csv')

rezoning_solar_data = rezoning_solar_data[rezoning_solar_data['ISO'].isin(iso_list)]

rezoning_windonshore_data = pd.read_csv('C:/Veda/VerveStacks/data/REZoning/REZoning_WindOnshore.csv')

rezoning_windonshore_data = rezoning_windonshore_data[rezoning_windonshore_data['ISO'].isin(iso_list)]
```

```python
"""
The ISO level shapes will be dervied by aggregating grid_cell level shapes for all grid cells needed to meet the demand of the ISO.
rezoning solar and wind data CFs will be updated grid_cell averages from atlite data.
atlite, rezoning solar and wind dfs - the three key ones are ready. iso_level shapes will be re-written with new data. It was never reflecting the aggregated values!
"""

# checking coverage of atlite data
# display(atlite_data)

avg_cf_solar_by_iso = atlite_data.groupby('grid_cell')['solar_capacity_factor'].mean()


avg_cf_windonshore_by_iso = atlite_data.groupby('grid_cell')['wind_capacity_factor'].mean()


# Add average solar capacity factor by grid_cell to the rezoning_solar_data DataFrame
# The mapping is from avg_cf_solar_by_iso (index: grid_cell) to rezoning_solar_data['grid_cell']
rezoning_solar_data['atlite_solar_cf'] = rezoning_solar_data['grid_cell'].map(avg_cf_solar_by_iso)

rezoning_solar_data.to_csv('REZoning_solar_with_atlite_cf.csv', index=False)

# Add average wind capacity factor by grid_cell to the rezoning_windonshore_data DataFrame
rezoning_windonshore_data['atlite_windonshore_cf'] = rezoning_windonshore_data['grid_cell'].map(avg_cf_windonshore_by_iso)

rezoning_windonshore_data.to_csv('REZoning_windonshore_with_atlite_cf.csv', index=False)
```
